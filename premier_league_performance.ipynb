```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('england-premier-league-teams-2018-to-2019-stats.csv')
df.set_index('league_position',inplace=True)
df.sort_index(inplace=True)

l = df['team_name'].map(lambda x : x[:3])

l[1] = 'ManC'
l[6] = 'ManU'
l = list(l)
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes

plt.plot(df.index,df['goals_scored_home'],'-.r',label='goals_scored_home')
plt.plot(df.index,df['goals_scored_away'],'-.g',label='goals_scored_away')

plt.plot(df.index,df['goals_conceded_home'],':m',label='goals_conceded_home')
plt.plot(df.index,df['goals_conceded_away'],':b',label='goals_conceded_away')

plt.plot(df.index,df['wins_home'],label='wins_home')
plt.plot(df.index,df['wins_away'],label='wins_away')


plt.legend()

ax.set_xticks(df.index)
ax.set_xticklabels(l)
ax.set_yticks(np.arange(0,65,10))
plt.xticks(rotation=90)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)



plt.subplots_adjust(bottom=0.5)
fig.autolayout : True
plt.title('Home vs Away performance in Premier League 18/19')
plt.grid()
plt.gcf().subplots_adjust(bottom=0.15)
plt.savefig('sort_prem.png',dpi=300, format='png', bbox_inches='tight')
```
